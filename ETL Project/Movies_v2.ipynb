```python
import requests
import pandas as pd

pd.options.display.max_rows = 100
```

```python
file = pd.read_csv('IMDB-Movie-Data.csv')

file.head()
```

# Clean Columns

```python
cleaned_file = file.rename(columns={'Rating': 'Score', 'Runtime (Minutes)': 'Runtime_minutes', 'Revenue (Millions)': 'Revenue_millions'})
cleaned_file.drop('Votes', axis=1, inplace=True)
cleaned_file['Rating'] = ''
cleaned_file.head()
```

```python
#test_data = renamed_file[:100].copy()
#test_data.head()
```

```python
for index, row in cleaned_file.iterrows():
    movie_name = row['Title']
    base_url = 'http://www.omdbapi.com/?i=tt3896198&apikey=7f03fa0f&'
    response = requests.get(f'{base_url}t={movie_name}').json()
    
    try:
        print(f'Retrieving rating data for {movie_name}')
        cleaned_file.loc[index, 'Rating'] = response['Rated']
    except(KeyError, IndexError):
        print('Missing data')
```
